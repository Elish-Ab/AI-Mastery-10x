```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```


Let's Upload data

```python
df = pd.read_csv('benin-malanville.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Visualize outliers using box plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']])
plt.title('Box Plot for Outlier Detection')
plt.show()
```

```python
# Example check for negative values
incorrect_entries = df[(df['GHI'] < 0) | (df['DNI'] < 0) | (df['DHI'] < 0)]
print("Incorrect entries:\n", incorrect_entries)
```

```python
df.describe(include=[np.number])
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)

plt.figure(figsize=(14, 7))
df[['GHI', 'DNI', 'DHI', 'Tamb']].resample('M').mean().plot()
plt.title('Monthly Trends of Solar Irradiance and Temperature')
plt.show()
```

```python
sns.lineplot(data=df, x='Timestamp', y='ModA', hue='Cleaning')
plt.title('Impact of Cleaning Events on ModA Sensor Readings')
plt.show()
```

```python
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.scatter(np.deg2rad(df['WD']), df['WS'])
ax.set_title('Polar Plot of Wind Direction and Speed')
plt.show()
```

```python
sns.scatterplot(data=df, x='RH', y='Tamb', hue='GHI')
plt.title('Relationship Between Humidity, Temperature, and GHI')
plt.show()
```

```python
df[['GHI', 'DNI', 'DHI', 'WS', 'Tamb']].hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Key Variables')
plt.show()
```

```python
df['GHI_zscore'] = stats.zscore(df['GHI'])
outliers = df[df['GHI_zscore'].abs() > 3]  # Flag values with z-score > 3
print("Outliers based on Z-score:\n", outliers)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GHI', y='Tamb', size='RH', hue='WS', sizes=(20, 200), alpha=0.5)
plt.title('Bubble Chart of GHI vs. Temperature vs. Wind Speed')
plt.show()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
# Removing rows with extreme Z-scores
df = df[df['GHI_zscore'].abs() <= 3]
```

```python
df.drop(columns=['Comments'], inplace=True)
```
